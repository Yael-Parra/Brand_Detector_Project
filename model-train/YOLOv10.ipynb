```python
from ultralytics import YOLO

# Cargar modelo base preentrenado (YOLOv10n)
model = YOLO("yolov10n.pt")

# Entrenar con tu dataset
results = model.train(
    data="../training-images/data.yaml",
    epochs=50,       # puedes ajustar según cantidad de datos
    imgsz=640,       # tamaño de imagen
    batch=16         # ajusta según RAM/GPU
)

# Validar el modelo
results = model.val()
```
