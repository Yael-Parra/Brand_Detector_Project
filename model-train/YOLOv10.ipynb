```python
from ultralytics import YOLO

# Cargar modelo base preentrenado (YOLOv10n)
model = YOLO("yolov10n.pt")

# Entrenar con tu dataset
results = model.train(
    data="../training-images/data.yaml",
    epochs=50,
    imgsz=640,
    batch=16
)

# Validar el modelo
results = model.val()
```

```python
# Detectar en una imagen nueva
results = model("../training-images/images/val/4.jpeg")
for r in results:
    r.show()
```
